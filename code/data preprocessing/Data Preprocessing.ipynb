```python
# 필요한 라이브러리 설치
!pip install pywavelets opencv-python matplotlib

# 라이브러리 임포트
import cv2
import numpy as np
import pywt
from matplotlib import pyplot as plt

# 이미지 불러오기 (컬러로 가져오기)
# 'your_image.jpg'를 처리하고자 하는 이미지 파일명으로 변경
image = cv2.imread('/content/00030.jpg', cv2.IMREAD_GRAYSCALE)

# 웨이블릿 변환 수행 (이산 웨이블릿 변환 - 2D)
coeffs2 = pywt.dwt2(image, 'haar')  # 'haar'는 해르 웨이블릿을 사용한다는 의미
LL, (LH, HL, HH) = coeffs2

# 변환된 서브밴드 LL만을 사용하여 복원 (단일 레벨 역 웨이블릿 변환)
coeffs2_reconstructed = LL, (np.zeros_like(LH), np.zeros_like(HL), np.zeros_like(HH))
reconstructed_image = pywt.idwt2(coeffs2_reconstructed, 'haar')

# 원본 이미지와 웨이블릿 변환 후 복원된 이미지를 비교
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# 원본 이미지
axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

# 복원된 이미지 (LL 성분만 사용)
axs[1].imshow(reconstructed_image, cmap='gray')
axs[1].set_title('Reconstructed Image (LL Band Only)')
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
# 필요한 라이브러리 설치
!pip install opencv-python matplotlib

# 라이브러리 임포트
import cv2
import numpy as np
from matplotlib import pyplot as plt

# 이미지 불러오기 (컬러로 가져오기)
# 'your_image.jpg'를 처리하고자 하는 이미지 파일명으로 변경
image = cv2.imread('/content/00030.jpg', cv2.IMREAD_COLOR)

# 가우시안 블러로 블러 처리된 이미지 생성
blurred = cv2.GaussianBlur(image, (9, 9), 10)

# 언샤프 마스크 적용: 원본 이미지와 블러 처리된 이미지의 차이로 샤프닝
sharp_image = cv2.addWeighted(image, 1.7, blurred, -0.5, 0)

# 원본 이미지와 샤프닝된 이미지를 비교
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# 원본 이미지
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image(Positive)', fontsize=16)
axs[0].axis('off')

# 언샤프 마스크 적용된 이미지
axs[1].imshow(cv2.cvtColor(sharp_image, cv2.COLOR_BGR2RGB))
axs[1].set_title('Gaussian Applied(Positive)', fontsize=16)
axs[1].axis('off')

plt.tight_layout()
plt.show()
```
