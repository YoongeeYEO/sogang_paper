# Domain Adaptation 적용 전 Source Domain과 Target Domain 이미지 pixel 분포 비교

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# 데이터셋의 이미지 파일 경로를 설정합니다.
dataset_path = '/content/ship dataset'

# 이미지 파일 경로 목록을 로드합니다.
image_paths = [os.path.join(dataset_path, fname) for fname in os.listdir(dataset_path) if fname.endswith('.jpg')]

# 이미지 픽셀 값을 저장할 리스트를 초기화합니다.
pixel_values = []

# 이미지 파일들을 반복하면서 픽셀 값을 로드합니다.
for img_path in image_paths:
    with Image.open(img_path) as img:
        img = img.resize((256, 256))  # 이미지 크기를 조정합니다.
        img_array = np.array(img)  # 이미지를 NumPy 배열로 변환합니다.
        pixel_values.extend(img_array.flatten())  # 픽셀 값을 리스트에 추가합니다.

# 픽셀 값의 분포를 히스토그램으로 시각화합니다.
plt.hist(pixel_values, bins=50, color='blue', alpha=0.7)
plt.title('Pixel Intensity Distribution')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
