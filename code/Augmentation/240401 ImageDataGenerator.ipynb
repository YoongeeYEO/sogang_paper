```python
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator


image = cv2.cvtColor(cv2.imread('/content/00027.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
def show_aug_image(image, generator, n_images=4):

    # ImageDataGenerator  image    4  .
    image_batch = np.expand_dims(image, axis=0)

    # featurewise_center or featurewise_std_normalization or zca_whitening  True fit 
    generator.fit(image_batch)
    # flow image batch generator .
    data_gen_iter = generator.flow(image_batch)

    fig, axs = plt.subplots(nrows=1, ncols=n_images, figsize=(24, 8))

    for i in range(n_images):
    	#generator batch size  augmentation (  )
        aug_image_batch = next(data_gen_iter)
        aug_image = np.squeeze(aug_image_batch)
        aug_image = aug_image.astype('int')
        axs[i].imshow(aug_image)
```

```python
import numpy as np

data_generator = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=(0.7, 1.3),
    horizontal_flip=True,
    vertical_flip=True,
    #rescale=1/255.0
)

show_aug_image(image, data_generator, n_images=4)
```

```python
data_generator = ImageDataGenerator(height_shift_range=0.3,
    width_shift_range=0.3,
    shear_range=45,
    zoom_range=0.2,
    rotation_range=90,
    channel_shift_range=50,
    brightness_range=[0.6, 1],
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='reflect'
)

show_aug_image(image, data_generator, n_images=4)
```

```python
data_generator = ImageDataGenerator(height_shift_range=0.5,
    width_shift_range=0.5,
    shear_range=60,
    zoom_range=0.4,
    rotation_range=120,
    channel_shift_range=50,
    brightness_range=[0.2, 1.3],
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='reflect'
)

show_aug_image(image, data_generator, n_images=4)
```
